```python
# multiple='stack' 
plt.figure(figsize=(10, 5))
sns.set_style('darkgrid')
sns.histplot(data=df, x='flipper_length_mm', hue='species')
plt.show()
```

```python
# multiple='stack' 
plt.figure(figsize=(10, 5))
sns.set_style('darkgrid')
sns.histplot(data=df, x='flipper_length_mm', hue='species', multiple='stack')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.set_style('whitegrid')
sns.kdeplot(data=df, x='flipper_length_mm', hue='species', multiple='stack')
plt.show()
```

```python
# 'col'         .
plt.figure(figsize=(10, 5))
sns.set_style('darkgrid')
sns.displot(data=df, x='flipper_length_mm', hue='species', col='species')
plt.show()
```

## Matplotlib    

* gridspec_kw 
    - fig, as = plt.subplot( , , gridspec_kw = dict (width_ratios = [a, b], heigh_ratios = [c, d]))
    - width_ratios   . width_ratios = [a, b] a:b .
    - heigh_ratios   . width_ratios = [c, d] c:d .

```python
fig, ax = plt.subplots(1, 2, figsize = (8,4), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=df, x='flipper_length_mm', y = 'bill_length_mm', hue='species', ax=ax[0])
sns.histplot(data=df, x='species', hue = 'species', shrink = .8, alpha=.6, legend=False, ax=ax[1])
fig.tight_layout()
```
