```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
from html_table_parser import parser_functions as parser
import numpy as np
```

```python
url="http://www.seoulmetro.co.kr/kr/page.do?menuIdx=366#none"
request=requests.get(url)
soup=BeautifulSoup(request.text,"html.parser")
```

#### [soup.select('span[class="Second_Tag"]')[0] select로 불러오는 법](https://blog.yonghyeon.com/44)

```python
#밑에 이걸 쓰면 user-select에 따라 달라지지 않는 페이지에서 원하는 표 추출하기 힘듦 
#data=soup.find("div",{"class":"tbl-box1 m-top20"})

#빈 리스트 하나 생성해서 데이터 프레임 담아주기
dfs=[]
for i in range(0,9):
#find랑 select랑 비슷한데 이거 쓰면 리스트 숫자만 바꿔서 전체 호선 출력 가능
    data=soup.select('div[class="tbl-box1 m-top20"]')[i]

    table=parser.make2d(data)
    df=pd.DataFrame(data=table[1:],columns=table[0])
    df['호선']=str(i+1)+"호선" #호선 칼럼 달아주기
    dfs.append(df)
```

#### [여러 개 데이터 테이블 for 문 참고페이지](https://hashcode.co.kr/questions/10662/%EB%B0%98%EB%B3%B5%EB%AC%B8%EC%9D%84-%EC%82%AC%EC%9A%A9%ED%95%B4%EC%84%9C-%EB%8D%B0%EC%9D%B4%ED%84%B0%ED%94%84%EB%A0%88%EC%9E%84%EC%9D%84-%EC%97%AC%EB%9F%AC%EA%B0%9C-%EB%A7%8C%EB%93%9C%EB%8A%94-%EB%B0%A9%EB%B2%95%EC%9D%B4-%EC%96%B4%EB%96%BB%EA%B2%8C-%EB%90%A0%EA%B9%8C%EC%9A%94)

#### [concat 참고 페이지](https://yganalyst.github.io/data_handling/Pd_12/#1-2-%EC%8B%9C%EB%A6%AC%EC%A6%88%EB%A5%BC-%EB%8D%B0%EC%9D%B4%ED%84%B0%ED%94%84%EB%A0%88%EC%9E%84%EC%97%90-%EB%B6%99%EC%9D%B4%EA%B8%B0)

```python
fini_df=pd.concat(dfs[0:9]) #파이썬에서 :ㅁ는  (ㅁ-1)까지
fini_df
```

```python
fini_df.reset_index(inplace=True) # 인덱스 재설정
fini_df.drop(columns=['index'], inplace=True) # 칼럼 삭제
fini_df
```

```python
fini_df.to_csv("서울교통공사 지하철 배려시설.csv",index=False,encoding = 'cp949') #csv로 내보내기
```
